### Questão 1: Cálculo da Soma

Dado o trecho de código abaixo, que soma os números inteiros de 1 a 13:



```python
def calcular_soma(n):
    return n * (n + 1) // 2

indice = 13
soma = calcular_soma(indice)
print(f"A soma dos primeiros {indice} números naturais é {soma}.")
```

#### Questão 2: Verificação na Sequência de Fibonacci



```python
import matplotlib.pyplot as plt

def calcula_fibonacci_ate(n):
    fib = [0, 1]
    while fib[-1] < n:
        fib.append(fib[-1] + fib[-2])
    return fib

def pertence_fibonacci(num, fib_sequence):
    return num in fib_sequence

numero = int(input("Informe um número: "))
fib_sequence = calcula_fibonacci_ate(numero)

if pertence_fibonacci(numero, fib_sequence):
    print(f"O número {numero} pertence à sequência de Fibonacci.")
else:
    print(f"O número {numero} não pertence à sequência de Fibonacci.")


```

```python
import json

# Carregar o arquivo JSON
with open('faturamento.json', 'r') as file:
    faturamento = json.load(file)

# Filtrar os valores de faturamento
valores = [dia['valor'] for dia in faturamento if dia['valor'] > 0]
dias = [dia['dia'] for dia in faturamento if dia['valor'] > 0]

# Cálculo do menor e maior valor de faturamento
menor_faturamento = min(valores)
maior_faturamento = max(valores)

# Cálculo da média mensal
media_mensal = sum(valores) / len(valores)

# Contagem de dias com faturamento acima da média
dias_acima_da_media = sum(1 for valor in valores if valor > media_mensal)

print(f"Menor Faturamento: R$ {menor_faturamento:.2f}")
print(f"Maior Faturamento: R$ {maior_faturamento:.2f}")
print(f"Dias com Faturamento Acima da Média: {dias_acima_da_media}")
```

```python
import matplotlib.pyplot as plt

# Gráfico de Faturamento Diário
plt.figure(figsize=(8, 4))
plt.plot(dias, valores, marker='o', linestyle='-', color='b')
plt.axhline(y=media_mensal, color='r', linestyle='--', label=f'Média Mensal: R$ {media_mensal:.2f}')
plt.xlabel('Dia')
plt.ylabel('Faturamento')
plt.title('Faturamento Diário')
plt.legend()
plt.grid(True)
plt.xticks(dias, rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Filtrando dias com faturamento acima da média
dias_acima = [dia for dia, valor in zip(dias, valores) if valor > media_mensal]
valores_acima = [valor for valor in valores if valor > media_mensal]

# Gráfico de Barras
plt.figure(figsize=(8, 4))
plt.bar(dias_acima, valores_acima, color='g')
plt.axhline(y=media_mensal, color='r', linestyle='--', label=f'Média Mensal: R$ {media_mensal:.1f}')
plt.xlabel('Dia')
plt.ylabel('Faturamento')
plt.title('Faturamento Acima da Média Mensal')
plt.legend()
plt.grid(True)
plt.xticks(dias, rotation=45)
plt.tight_layout()
plt.show()
```

#### Questão 4: Percentual de Representação por Estado

```python


faturamento = {
    'SP': 67836.43,
    'RJ': 36678.66,
    'MG': 29229.88,
    'ES': 27165.48,
    'Outros': 19849.53
}

total = sum(faturamento.values())
percentuais = {estado: (valor / total) * 100 for estado, valor in faturamento.items()}

for estado, percentual in percentuais.items():
    print(f"{estado}: {percentual:.2f}%")

# Visualização dos resultados
plt.figure(figsize=(8, 8))
plt.pie(percentuais.values(), labels=percentuais.keys(), autopct='%1.1f%%', startangle=140)
plt.title('Percentual de Faturamento por Estado')
plt.show()
```

#### Questão 5: Inversão de String


```python
def inverter_string(s):
    n = len(s)
    invertida = [''] * n
    for i in range(n):
        invertida[n - i - 1] = s[i]
    return ''.join(invertida)

string = input("Informe uma string: ")
print(f"String invertida: {inverter_string(string)}")
```
